```python
US_BUSINESS_DAY = CustomBusinessDay(calendar=USFederalHolidayCalendar())
```

```python
Tickers_EarningsCallsDates = pd.read_csv("../data/Descending_Ticker_Industry.csv")
Tickers_EarningsCallsDates
```

```python
Tickers_EarningsCallsDates['DayBefore_EarningsCall'] = Tickers_EarningsCallsDates['date_earningscall'] - 1*US_BUSINESS_DAY
# Tickers_EarningsCallsDates['ClosingPrice_DayBefore_EarningsCall'] = ""
Tickers_EarningsCallsDates['DateOf_EarningsCall'] = Tickers_EarningsCallsDates['date_earningscall'] 
# Tickers_EarningsCallsDates['ClosingPrice_DayOf_EarningsCall'] = ""
Tickers_EarningsCallsDates['DayAfter_EarningsCall'] = Tickers_EarningsCallsDates['date_earningscall'] + 1*US_BUSINESS_DAY 
# Tickers_EarningsCallsDates['ClosingPrice_DayAfter_EarningsCall'] = ""
# Tickers_EarningsCallsDates['PercentChange_bt_ClosingPrices_DayBefore_y_DayAfter'] = ""
```
